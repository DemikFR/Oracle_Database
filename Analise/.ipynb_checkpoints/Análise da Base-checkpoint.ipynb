```python
df.isnull().any()
```

### Inicio da bolsa não pode ser nula, por isso, deverá ser verificado se no campo COTA, vale como inicio.

####  Verificar quantos registros correspondem a hipótese acima

```python
df[df['cota'].str.extract(r'(20[0-9]{2})', expand=False) == pd.to_datetime(df['inicio']).dt.strftime('%Y')].shape[0]
```

#### Para entender o motivo dos outros 7292 não serem iguais

```python
df1 = df.loc[df['cota'].str.extract(r'(20[0-9]{2})', expand=False) != pd.to_datetime(df['inicio']).dt.strftime('%Y'), 
             ['cota', 'inicio', 'fim']]
df1
```

#### Possibilidade: Primeiro ano da cota foi diminuido em 1, pelo motivo de ter finalizado no mesmo ano, ** Confirmar isso **

#### Existem algumas colunas que o Regex está trazendo nulo, pois tem apenas texto, por isso está entre as colunas que divergeram a hipótese

```python
df1 = df1.loc[df['cota'].str.extract(r'(20[0-9]{2})', expand=False).notnull()]
df1
```

#### Tirar os valores nulos de inicio, pois não poderão ser tratados.

```python
df1 = df1.loc[pd.to_datetime(df1['inicio']).dt.strftime('%Y').notnull()]
df1
```

#### Buscar os registros em que o ano de cota (inicial) não bate com o campo inicio

```python
df1.loc[df1['cota'].str.extract(r'(20[0-9]{2})', expand=False).astype(int)+1 != pd.to_datetime(df1['inicio']).dt.strftime('%Y')
        .astype(int)]
```

#### Nota-se, que há datas exageradas, invertidas (inicio no fim e vice-versa) e de apenas um ano, para os outros casos, possivelmente segue a mesma linha da 1° hipótese, porém foi aumentado ou diminuido o ano inicial, pelo motivo da data de inicio ter sido a mesma do ano fim.

#### Para finalizar essa hipótese, deve-se verificar se tem um padrão de diminuição ou aumento desses anos.

```python
# Para as que diminuem o ano

df2 = df1[df1['cota'].str.extract(r'(20[0-9]{2})', expand=False).astype(int)+1 == pd.to_datetime(df1['inicio']).dt
          .strftime('%Y').astype(int)]
pd.DatetimeIndex(df2['inicio']).month.max()
```

```python
# Para as que aumentam o ano

df3 = df1[df1['cota'].str.extract(r'(20[0-9]{2})', expand=False).astype(int)+1 != pd.to_datetime(df1['inicio']).dt
          .strftime('%Y').astype(int)]
pd.DatetimeIndex(df3['inicio']).month.min()
```

#### Não seguem um padrão por mês, então, esses não deverão ser consideradas como iguais no tratamento, apenas se no caso do inicio, que é obrigatório

#### Verificar se os registros que tem apenas 1 ano, são Inicio ou fim

```python
df5 = df1[(~df1['cota'].str.contains(r'(20[0-9]{2}-20[0-9]{2})')) & (~df1['cota'].str.contains(r'(-)'))]
df5
```

```python
df5[df5['cota'] == df5['inicio'].dt.strftime('%Y')].count()
```

```python
df5[df5['cota'] == df5['fim'].dt.strftime('%Y')].count()
```

### Não seguem um padrão, porém poderá ser usado para tratar os dados nulos, usando a lógica de se o ano de cota for maior que inicio, ele é fim, e vice-versa. Nos invertidos, deverá ser tratado para entrar no banco.

### No campo codigo_projeto, existe a possibilidade de uma redundância com o campo "ano", pois depois do traço, também contém o ano de projeto.

```python
df6 = df[df['codigo_projeto'].str.extract(r'(20[0-9]{2})', expand=False) != df['ano'].astype(int)]
df6
```
